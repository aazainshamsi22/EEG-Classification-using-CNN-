```python
import mne
import pandas as pd

# Load the GDF dataset (replace 'dataset.gdf' with the actual filename and path)
raw = mne.io.read_raw_gdf('/content/drive/MyDrive/BCICIV_2a_gdf/A01E.gdf', preload=True)

# Extract the data and labels from the GDF dataset
data = raw.get_data()
labels = raw.annotations.description

# Handle mismatch in data and labels lengths
if len(labels) < data.shape[1]:
    labels = list(labels) + ['Unknown'] * (data.shape[1] - len(labels))

# Create a DataFrame with the data and labels
df = pd.DataFrame(data.T, columns=[f'Channel_{i+1}' for i in range(data.shape[0])])
df['Label'] = labels

# Save the DataFrame as a CSV file
df.to_csv('dataset.csv', index=False)
```

```python
# Drop the last 3 channels from the DataFrame
df_preprocessed = df.drop(df.columns[-4:], axis=1)

# Verify the preprocessed DataFrame
print(df_preprocessed.head())

# Save the preprocessed DataFrame as a CSV file
df_preprocessed.to_csv('preprocessed_dataset.csv', index=False)
```
